__1. How old were the youngest male and female participants of the 1996 Olympics?__

- 16 and 15
- **14 and 12**
- 16 and 12
- 13 and 11

```python
data.loc[data.Games == '1996 Summer'].groupby('Sex').Age.agg(min)
```

__2. What was the percentage of male gymnasts among all the male participants of the 2000 Olympics? Round the answer to the first decimal.__

*Hint:* here and further if needed drop duplicated sportsmen to count only unique ones. 

- 0.2
- **1.5**
- 2.5
- 7.7

```python
q2 = pd.DataFrame(
    (data
      .loc[data.Year == 2000]
      .groupby(['Sport', 'Sex'])
      .Name
      .nunique()
     ))
q2.rename(columns = {'Name' : 'Obs'}, inplace = True)
#q2.head()
ans2 = q2.query("Sport == 'Gymnastics' & Sex == 'M'").Obs.sum() / q2.query("Sex == 'M'").Obs.sum()
ans2 = (100 * ans2).round(1)
print(ans2)
```

__3. What are the mean and standard deviation of height for female basketball players participated in the 2000 Olympics? Round the answer to the first decimal.__

- 178.5 and 7.2
- 179.4 and 10
- 180.7 and 6.7
- **182.4 and 9.1** 

```python
q3 = pd.DataFrame(
    (data
    .loc[(data.Year == 2000) & (data.Sex == 'F') & (data.Sport == 'Basketball')]
    .drop_duplicates()
    .Height
    .agg(['mean', 'std'])
    .round(1)))
q3
```

__6. How many silver medals in tennis did Australia win at the 2000 Olympics?__

- 0

```python
q6 = (data
     .loc[(data.Year == 2000) & 
          (data.Team == 'Australia') & 
          (data.Medal == 'Silver') & 
          (data.Sport == 'Tennis')]
      )
q6.Medal.value_counts()
```

__7. Is it true that Switzerland won fewer medals than Serbia at the 2016 Olympics? Do not consider NaN values in _Medal_ column.__ 

- **Yes**
- No

```python
q7_sw = (data
        .loc[(data.Year == 2016) &
             (data.Team == 'Switzerland')]
        .dropna(subset = ['Medal'])
        .Medal
        .value_counts()        
        )
#q7_sw
q7_se = (data
        .loc[(data.Year == 2016) &
             (data.Team == 'Serbia')]
        .dropna(subset = ['Medal'])
        .Medal
        .value_counts()        
        )
#q7_se
q7_sw.sum() < q7_se.sum()
```

__8. What age category did the fewest and the most participants of the 2014 Olympics belong to?__
- [45-55] and [25-35) correspondingly
- **[45-55] and [15-25) correspondingly**
- [35-45] and [25-35) correspondingly
- [45-55] and [35-45) correspondingly

```python
q8 = (data.loc[data.Year == 2014])
# q8.Age.min() -- minimum is 15 thereby I will create bins starting from 15
# q8.Age.max() -- maximum is 55 thereby I will create bins upto 55
q8['age_bins'] = pd.cut(x = q8.Age, bins = [15, 25, 35, 45, 55])
q8.age_bins.value_counts()
```

__9. Is it true that there were Summer Olympics held in Lake Placid? Is it true that there were Winter Olympics held in Sankt Moritz?__

- Yes, Yes

- **No, Yes**

__10. What is the absolute difference between the number of unique sports at the 1995 Olympics and 2016 Olympics?__

- 16
- 24
- 26
- **34**

```python
sports_95 = (data
            .loc[data.Year == 1995]
            .Sport
            .nunique())
sports_16 = (data
            .loc[data.Year == 2016]
            .Sport
            .nunique())
abs(sports_95 - sports_16)
```
